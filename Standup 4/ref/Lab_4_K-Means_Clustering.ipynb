# Principle Component Analysis:
PCA allows us to capture more information in our analysis by capturing relevant variables
Center the data to have a mean of zero and standard deviation 1
Variability of observations along a given dimension is of interest
Using correlated variables in the analysis will not help much on the one hand and can become computationally expensive on the other hand when number of features (p) is large and several observations exist 
PCA will provide a low-dimensional representation that accounts for a lot of the information 
The data can then be plotted in the computed low dimensional space using the components which best explain the variance measured by the scores
Constrain the loadings to ensure the assume of squares is = 1

Below you will find code sourced from various sources through Google.com. You are encouraged to reference it and suitably adjust for your needs. 

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
%matplotlib inline
from sklearn import datasets

# Boston Dataset
boston = datasets.load_boston()
X = boston.data

#KMeans    
km = KMeans(n_clusters=3,
           n_init = 10,
           max_iter=2000,
           tol=1e-4)
km.fit(X)
km.predict(X)
labels = km.labels_


#Plotting
fig = plt.figure(1, figsize=(7,7))
ax = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)
ax.scatter(X[:, 0], X[:, 5], X[:, 12],
          c=labels.astype(np.float), edgecolor="k", s=50)
ax.set_xlabel("CRIM")
ax.set_ylabel("RM")
ax.set_zlabel("LSTAT")
plt.title("K Means", fontsize=14)
```

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Iris Dataset
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris['feature_names'])
#print(X)
data = X[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)']]

sse = {}
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(data)
    data["clusters"] = kmeans.labels_
    #print(data["clusters"])
    sse[k] = kmeans.inertia_ # Inertia: Sum of distances of samples to their closest cluster center
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```
